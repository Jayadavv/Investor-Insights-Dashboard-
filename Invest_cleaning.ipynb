```python
# Most Preferred Investment Avenue
preferred_investment = df['Avenue'].value_counts().idxmax()
print(f'Most preferred investment avenue: {preferred_investment}')
```

```python
# Reasons for Investment
selected_columns = df.loc[:, 'What are your savings objectives?':'Reason_FD']

combined_reasons = selected_columns.apply(lambda row: ' '.join(row.values.astype(str)), axis=1)


from collections import Counter

word_counts = Counter(" ".join(combined_reasons).split())


recurring_reasons = word_counts.most_common(10)
print("Top recurring reasons:")
for reason, count in recurring_reasons:
    print(f"{reason}: {count}")
```

```python
# Investment Duration

# Defining the mapping for ranges to numeric values
duration_mapping = {
    'less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'more than 5 years': 6
}

# Mapping the values to the 'Duration' column
df['Investment Duration (Numeric)'] = df['Duration'].map(duration_mapping)

# Calculate the mean
avg_duration = df['Investment Duration (Numeric)'].mean()
print(f'Average investment duration (in years): {avg_duration:.2f}')

# Determine the most frequent range
most_frequent = df['Duration'].value_counts().idxmax()
print(f'Most frequent investment duration range: {most_frequent}')
```

```python
expectations = df['Expect'].value_counts()
expectations.reset_index()
```

```python
# Extract relevant columns: age, Duration, and Expect
# Handle categorical columns by encoding them into numerical values
data_selected = df[['age', 'Duration', 'Expect']]

# Encoding 'Duration' and 'Expect' as they are categorical
# Example mappings: 
# 'Duration': {'Less than 1 year': 0, '1-3 years': 1, '3-5 years': 2, 'More than 5 years': 3}
# 'Expect': {'10%-20%': 0.15, '20%-30%': 0.25, ...}

duration_mapping = {'Less than 1 year': 0, '1-3 years': 1, '3-5 years': 2, 'More than 5 years': 3}
expect_mapping = {'10%-20%': 0.15, '20%-30%': 0.25, '30%-40%': 0.35, '40%-50%': 0.45}

data_selected['Duration'] = data_selected['Duration'].map(duration_mapping)
data_selected['Expect'] = data_selected['Expect'].map(expect_mapping)

# Drop rows with missing values after encoding
data_selected = data_selected.dropna()

# Calculate correlation matrix
correlation_matrix = data_selected.corr()

# Visualize the correlation matrix using seaborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Selected Features')
plt.show()

correlation_matrix
```
